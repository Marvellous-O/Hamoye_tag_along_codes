```python
#Importing the supply dataset
df_= pd.read_csv('/content/Africa Food Supply (2004 - 2013).csv')
df_.head()
```

```python
#Change the Year column to Datetime
df_['Year']=pd.to_datetime(df_.Year, format='%Y').dt.year
df_.head()
```

```python
#To check the data types
df_.info
```

```python
#adding the unit to the value column
df_['Value_2']= df_['Value'].astype(str)
df_['Value_2']=df_['Value_2']+ 'kcal/person/day'
df_.head()
```

```python
#Grouping by Year to show sum of values
df1=df_.groupby('Year')['Value'].sum().reset_index()
df1.head
```

```python
#sns.boxplot(data=df_, x="Country", y="Value")
df_2012 = df_[df_['Year'] == 2012]
sns.boxplot(data =df_2012, x='Country' ,y='Value')
```

```python
#Plotting a boxplot to show the values by each year revealing the outliers
sns.boxplot(data=df_, x="Year", y="Value")
```

```python
#Boxplot of Values for Each Year for the Production dataset
plt.figure(figsize=(8, 4))
plt.boxplot([df[df['Year'] == year]['Value'] for year in df['Year'].unique()],
            labels=df['Year'].unique())

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Boxplot of Values for Each Year')

# Show the plot
plt.show()
```

```python
#Boxplot of Values for Each Year for the Supply dataset
plt.figure(figsize=(8, 4))
plt.boxplot([df_[df_['Year'] == year]['Value'] for year in df_['Year'].unique()],
            labels=df_['Year'].unique())

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Boxplot of Values for Each Year')

# Show the plot
plt.show()
```

```python
#box plot showing for year 2004 only with multiple outliers
selected_year = 2004

# Filter the data for the selected year
selected_year_data = df_[df_['Year'] == selected_year]

# Create a boxplot for the selected year
plt.figure(figsize=(6, 4))
plt.boxplot(selected_year_data['Value'], labels=[selected_year])

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Value')
plt.title(f'Boxplot of Values for 2004')

# Show the plot
plt.show()
```

```python
# boxplot showing the values grouped by countries
plt.figure(figsize=(10, 6))
boxplot = df_.boxplot(by='Country', column='Value', grid=False)

# Add labels and title
plt.xlabel('Country')
plt.ylabel('Value')
plt.title('Boxplot of Values by Country')
plt.show()
```
